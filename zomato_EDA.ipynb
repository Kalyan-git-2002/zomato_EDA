```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head(5)
```

```python
final_ds = pd.merge(df,df_country, on = 'Country Code' , how = 'left') # group the two dataframes
final_ds
```

```python
final_ds.dtypes # shows data types in final dataset
```

```python
final_ds.columns
```

```python
Country_Names = final_ds.Country.value_counts().index #gives index on which value count work
Country_Names
```

```python
Count_country = final_ds.Country.value_counts().values # gives value count as array
Count_country
```

```python
#piechart - top 3 countries using zomato
plt.pie(Count_country[:3],labels = Country_Names[:3],autopct = '%1.0f%%')
```

```python
# Analysis of Rating
Grouped_data = final_ds.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:"Rating_count"})
Grouped_data.head(5)
```

```python
plt.rcParams['figure.figsize'] = (16,4)
sns.barplot(data = Grouped_data,x = 'Aggregate rating',y = 'Rating_count',hue='Rating color',
           palette = {'Red':'Red','White' : 'Blue','Orange':'Orange','Green':'Green','Dark Green':'darkgreen','Yellow':'Yellow'})
```

```python
pie_data = final_ds.groupby(['Rating text','Rating color']).size().reset_index().rename(columns = {0:"Rating_count"})
custom_order = ['Excellent' , 'Very Good' ,'Good' , 'Average', 'Poor','Not rated']
pie_data['Rating text'] = pd.Categorical(pie_data['Rating text'],categories = custom_order,ordered = True)
pie_data = pie_data.sort_values('Rating text')
pie_data
```

```python
plt.rcParams['figure.figsize'] = (8,8)
c_map = {'Red':'Red','White' : 'Blue','Orange':'Orange','Green':'Green','Dark Green':'darkgreen','Yellow':'Yellow'}
plot_color = pie_data['Rating color'].map(c_map).tolist()
plt.pie(pie_data['Rating_count'],labels = pie_data['Rating text'],colors = plot_color , autopct = '%1.0f%%',startangle = 30)
plt.show()
```

Observations:
1. NOT RATED count is very high.
2. High ratings are average.

```python
## Find the countries that given zero rating.

final_ds[final_ds['Aggregate rating'] == 0.0].groupby(['Country']).size().reset_index()
```

Observations :
 More '0' ratings from indian customers

```python
## find out which country use which Currency

final_ds[['Country','Currency']].groupby(['Country','Currency']).sum().reset_index()
```

observation :
only india and UAE has online delivery
